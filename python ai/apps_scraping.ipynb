```python
import requests
from bs4 import BeautifulSoup

url = "https://appfigures.com/top-apps/ios-app-store/united-states/iphone/top-overall"

headers = {    
    "User-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36",
    "Referer":"https://appfigures.com/top-apps/ios-app-store/united-states/iphone/top-overall",
    "Accept-language":"en-US,en;q=0.9",
}

responce = requests.get(url, headers=headers, timeout=10)
print(responce.status_code)
soup = BeautifulSoup(responce.text, "lxml")
```
